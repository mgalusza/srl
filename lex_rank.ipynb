```python
def clearTags(tags):
    ret = []
    for t in tags:
        if t == 'B-V': 
            ret.append(t)
            continue
        if t[0] == 'O'  : 
            ret.append(t)
            continue
        if t[2] == 'A': 
            ret.append(t)
            continue
        else:
            ret.append('O')
            continue
    return ret


def clearTag(t):
    ret = []
    if t == 'B-V': 
        return 'V'
    elif t[0] == 'O': 
        return 'O'
    elif t[2] == 'A': 
        return  t[2:]
    else:
        return 'O'

    
def tagOverlay(src,k):
    # overlay tags starting from k to number of verbs
    ret = [1 for i in range(0,len(src[k][1]))]
    src_tags = src[k][1]
    for j in range(k+1,len(src)): # ovelay frames following k
        for i in range(0,len(src[j][1])): # get tags for k-th frame
            if src[j][1][i] == 'O':
                ret[i] = src_tags[i]
            else:
                ret[i] = 'U'
        src_tags = ret
    return ret
```

```python
def frame2nodes(parsed):
    nodes = []
    pos   = []
    lemma = []
    overlay = []
    next_verb = []
    frame = ""
    arg   = ""
    hasVerb = False
    current_node_type = 'START'
    if parsed == []:
        return []
    for w in parsed:
        #detect frame
        if w[4] != '':
            frame = frame + " " + w[3]
    for w in parsed:  
        if w[6] == 'VERB':
            if w[4]!= '' and  w[4] != 'V':
                 hasVerb = True
        if w[4] == current_node_type:
            arg = arg + " " + w[3]
            lemma.append(w[5])
            pos.append(w[6])
            overlay.append(w[7])
            next_verb.append(w[8])
            
        if current_node_type == 'START':
            if w[4] != '':
                current_node_type = w[4]
                pos = [w[6]]
                lemma = [w[5]]
                overlay = [w[7]]
                next_verb = [w[8]]
                arg = arg + " " + w[3]
        if w[4] != current_node_type:
            if w[4] != "":
                nodes.append((arg,current_node_type,frame,hasVerb,w[0],pos,lemma,overlay,next_verb))
                arg = w[3] 
                pos = [w[6]]
                lemma = [w[5]]
                overlay = [w[7]]
                next_verb= [w[8]]
                current_node_type = w[4]
                hasVerb = False
    nodes.append((arg,current_node_type,frame,hasVerb,w[0],pos,lemma,overlay,next_verb))
    return nodes
```

```python
import hashlib

def node2graph(G,nodes):
    for n in nodes:
        if n[1] == 'V':
            verb_node = (n[0].strip(),'verb',n[4])
            frame_d = n[2].strip()
            if frame_d == '':
                frame_d = 'E'
                result = '-1'
                
            else:
                encoded = frame_d.encode()
                result = hashlib.sha256(encoded)
            G.add_node(verb_node,frame=frame_d,frame_id = result.hexdigest(),centering_nouns = ())
    for n in nodes:       
        if n[1] != 'V':      
            if n[1][0] == 'A':
                node = (n[0],'role',n[4])
                G.add_node(node, frame = frame_d, frame_id = result.hexdigest(), role=n[1])
                G.add_edge(verb_node,node,weight = 1,role=n[1])
                if n[1] in ['ARG0','ARG1']:
                    t = G.nodes[verb_node]['centering_nouns']
                    nouns = set()
                    for i in range(0,len(n[5])):
                        
                        if n[5][i] in ['NOUN','PROPN']:

                            nouns.add(n[6][i])
                            
                    G.nodes[verb_node]['centering_nouns'] = t + tuple(nouns)
                for i in range(0,len(n[5])):
                    if n[5][i] == 'VERB':
                        verb = n[0].strip().split(" ")[i]
                        if verb == n[8][i]:
                            verb_1 = (verb,'verb',n[4])
                            if G.has_node(verb_1) == False:
                                G.add_node(verb_1,frame = 'E',frame_id = '-1')
                            G.add_edge(node,verb_1,weight = 1,role='verb')
                #else:
                  #  for i in range(0,len(n[5])):
                    if n[5][i] in ['NOUN','PROPN'] and n[7][i] !='U':
                        noun = n[6][i]
                        G.add_node((noun,'noun'),frame_id = '-1', frame = 'EMPTY')
                        G.add_edge(node,(noun,'noun'),weight = 1,role='noun')
    return G
```

```python
def findPaths(G,u,n,end_node,excludeSet = None):
    if excludeSet == None:
        excludeSet = set([u])
    else:
        excludeSet.add(u)
    if n==0:
        return [[u]]
    if u==end_node:
        return [[u]]
    paths = [[u]+path for neighbor in G.neighbors(u) if neighbor not in excludeSet for path in findPaths(G,neighbor,n-1,end_node,excludeSet)]
    excludeSet.remove(u)
    return paths
```

```python
def findPathTailRecursive(G,u,end_node,n,excludeSet=None,path=[]):
    if excludeSet == None:
        excludeSet = set([u])
        path = [u]
    else:
        excludeSet.add(u)
    if end_node == None and n==0:
        paths.append(path)
        return path
    if (end_node != None and u==end_node): 
        paths.append(path)
        return path
    if (end_node != None and u!=end_node) and n==0: 
        #paths.append(path)
        return path
    for neighbor in G.neighbors(u):
        if neighbor not in excludeSet:
            findPathTailRecursive(G,neighbor,end_node,n-1,excludeSet,path + [neighbor])
    excludeSet.remove(u)
            
def run_findPathTailRecursive(G,u,end_node,n):
    global paths
    paths = []
    findPathTailRecursive(G,u,end_node,n)
    return paths
```

```python
# Define a new example sentence with multiple meanings of the word "bank"
text = "After stealing money from the bank vault, the bank robber was seen " \
       "fishing on the Mississippi river bank."

# Add the special tokens.
marked_text = "[CLS] " + text + " [SEP]"

# Split the sentence into tokens.
tokenized_text = tokenizer.tokenize(marked_text)

# Map the token strings to their vocabulary indeces.
indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text)

# Display the words with their indeces.
for tup in zip(tokenized_text, indexed_tokens):
    print('{:<12} {:>6,}'.format(tup[0], tup[1]))
```

```python
# Mark each of the 22 tokens as belonging to sentence "1".
segments_ids = [1] * len(tokenized_text)

print (segments_ids)
```

```python
# Convert inputs to PyTorch tensors
tokens_tensor = torch.tensor([indexed_tokens])
segments_tensors = torch.tensor([segments_ids])

# Load pre-trained model (weights)
model = BertModel.from_pretrained('bert-base-uncased')

# Put the model in "evaluation" mode, meaning feed-forward operation.
model.eval()
```

```python
# Predict hidden states features for each layer
with torch.no_grad():
    encoded_layers, _ = model(tokens_tensor, segments_tensors)
```

```python
print ("Number of layers:", len(encoded_layers))
layer_i = 0

print ("Number of batches:", len(encoded_layers[layer_i]))
batch_i = 0

print ("Number of tokens:", len(encoded_layers[layer_i][batch_i]))
token_i = 0

print ("Number of hidden units:", len(encoded_layers[layer_i][batch_i][token_i]))
```

```python
# For the 5th token in our sentence, select its feature values from layer 5.
token_i = 5
layer_i = 5
vec = encoded_layers[layer_i][batch_i][token_i]

# Plot the values as a histogram to show their distribution.
plt.figure(figsize=(10,10))
plt.hist(vec, bins=200)
plt.show()
```

```python
# `encoded_layers` is a Python list.
print('     Type of encoded_layers: ', type(encoded_layers))

# Each layer in the list is a torch tensor.
print('Tensor shape for each layer: ', encoded_layers[0].size())
```

```python
# Concatenate the tensors for all layers. We use `stack` here to
# create a new dimension in the tensor.
token_embeddings = torch.stack(encoded_layers, dim=0)

token_embeddings.size()
```

```python
# Remove dimension 1, the "batches".
token_embeddings = torch.squeeze(token_embeddings, dim=1)

token_embeddings.size()
```

```python
# Swap dimensions 0 and 1.
token_embeddings = token_embeddings.permute(1,0,2)

token_embeddings.size()
```

```python
# `encoded_layers` has shape [12 x 1 x 22 x 768]

# `token_vecs` is a tensor with shape [22 x 768]
token_vecs = encoded_layers[11][0]

# Calculate the average of all 22 token vectors.
sentence_embedding = torch.mean(token_vecs, dim=0)
```
