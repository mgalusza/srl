```python
import nltk
from nltk.corpus import brown
from nltk import word_tokenize, pos_tag, sent_tokenize

# Load the Brown Corpus
corpus = brown

# Initialize variables for total verbs and total sentences
total_verbs = 0
total_sentences = 0
verbs_cnt_arr = []
max_verb_count = float("-inf")  # Initialize with negative infinity
min_verb_count = float("inf")   # Initialize with positive infinity

# Loop through the categories (genres) in the Brown Corpus
for category in corpus.categories():
    # Get sentences from the corpus for the current category
    sentences = corpus.sents(categories=category)
    total_sentences += len(sentences)

    # Tokenize and tag each sentence, then count verbs
    for sentence in sentences:
        words = word_tokenize(' '.join(sentence))
        tags = pos_tag(words)
        verb_count = sum(1 for word, tag in tags if tag.startswith('VB'))
        verbs_cnt_arr.append(verb_count)
        total_verbs += verb_count
        if verb_count > max_verb_count:
            max_verb_count = verb_count
        if verb_count < min_verb_count:
            min_verb_count = verb_count
# Calculate the average number of verbs per sentence
average_verbs_per_sentence = total_verbs / total_sentences

print(f"Average number of verbs per sentence in the Brown Corpus: {average_verbs_per_sentence:.2f}")
print(f"Minumum of verbs per sentence: {min_verb_count}")
print(f"Max number of verbs per sentence: {max_verb_count}")
```

```python
def clearTags(tags):
    ret = []
    for t in tags:
        if t == 'B-V': 
            ret.append(t)
            continue
        if t[0] == 'O'  : 
            ret.append(t)
            continue
        if t[2] == 'A': 
            ret.append(t)
            continue
        else:
            ret.append('O')
            continue
    return ret


def clearTag(t):
    ret = []
    if t == 'B-V': 
        return 'V'
    elif t[0] == 'O': 
        return 'O'
    elif t[2] == 'A': 
        return  t[2:]
    else:
        return 'O'

    
def tagOverlay(src,k):
    # overlay tags starting from k to number of verbs
    ret = [1 for i in range(0,len(src[k][1]))]
    src_tags = src[k][1]
    for j in range(k+1,len(src)): # ovelay frames following k
        for i in range(0,len(src[j][1])): # get tags for k-th frame
            if src[j][1][i] == 'O':
                ret[i] = src_tags[i]
            else:
                ret[i] = 'U'
        src_tags = ret
    return ret
```

```python
import torch
from pytorch_pretrained_bert import BertTokenizer, BertModel, BertForMaskedLM

# OPTIONAL: if you want to have more information on what's happening, activate the logger as follows
import logging
#logging.basicConfig(level=logging.INFO)

import matplotlib.pyplot as plt


# Load pre-trained model tokenizer (vocabulary)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
text = "Here is the sentence I want embeddings for."
marked_text = "[CLS] " + text + " [SEP]"

# Tokenize our sentence with the BERT tokenizer.
tokenized_text = tokenizer.tokenize(marked_text)

# Print out the tokens.
print (tokenized_text)
```

```python
# Define a new example sentence with multiple meanings of the word "bank"
text = "After stealing money from the bank vault, the bank robber was seen " \
       "fishing on the Mississippi river bank."

# Add the special tokens.
marked_text = "[CLS] " + text + " [SEP]"

# Split the sentence into tokens.
tokenized_text = tokenizer.tokenize(marked_text)

# Map the token strings to their vocabulary indeces.
indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text)

# Display the words with their indeces.
for tup in zip(tokenized_text, indexed_tokens):
    print('{:<12} {:>6,}'.format(tup[0], tup[1]))
```

```python
# Mark each of the 22 tokens as belonging to sentence "1".
segments_ids = [1] * len(tokenized_text)

print (segments_ids)
```

```python
# Convert inputs to PyTorch tensors
tokens_tensor = torch.tensor([indexed_tokens])
segments_tensors = torch.tensor([segments_ids])

# Load pre-trained model (weights)
model = BertModel.from_pretrained('bert-base-uncased')

# Put the model in "evaluation" mode, meaning feed-forward operation.
model.eval()
```

```python
# Predict hidden states features for each layer
with torch.no_grad():
    encoded_layers, _ = model(tokens_tensor, segments_tensors)
```

```python
print ("Number of layers:", len(encoded_layers))
layer_i = 0

print ("Number of batches:", len(encoded_layers[layer_i]))
batch_i = 0

print ("Number of tokens:", len(encoded_layers[layer_i][batch_i]))
token_i = 0

print ("Number of hidden units:", len(encoded_layers[layer_i][batch_i][token_i]))
```

```python
# For the 5th token in our sentence, select its feature values from layer 5.
token_i = 5
layer_i = 5
vec = encoded_layers[layer_i][batch_i][token_i]

# Plot the values as a histogram to show their distribution.
plt.figure(figsize=(10,10))
plt.hist(vec, bins=200)
plt.show()
```

```python
# `encoded_layers` is a Python list.
print('     Type of encoded_layers: ', type(encoded_layers))

# Each layer in the list is a torch tensor.
print('Tensor shape for each layer: ', encoded_layers[0].size())
```

```python
# Concatenate the tensors for all layers. We use `stack` here to
# create a new dimension in the tensor.
token_embeddings = torch.stack(encoded_layers, dim=0)

token_embeddings.size()
```

```python
# Remove dimension 1, the "batches".
token_embeddings = torch.squeeze(token_embeddings, dim=1)

token_embeddings.size()
```

```python
# Swap dimensions 0 and 1.
token_embeddings = token_embeddings.permute(1,0,2)

token_embeddings.size()
```

```python
# `encoded_layers` has shape [12 x 1 x 22 x 768]

# `token_vecs` is a tensor with shape [22 x 768]
token_vecs = encoded_layers[11][0]

# Calculate the average of all 22 token vectors.
sentence_embedding = torch.mean(token_vecs, dim=0)
```
