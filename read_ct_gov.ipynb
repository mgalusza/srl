```python
import requests

def search_clinical_trials(search_term, max_results=10):
    base_url = "https://ClinicalTrials.gov/api/query/study_fields"
    params = {
        "expr": search_term,
        "min_rnk": 1,
        "max_rnk": max_results,
        "fields": "NCTId,Phase,StartDate,StudyType,Condition,BriefTitle,EligibilityCriteria",
        "fmt": "json",
    }

    response = requests.get(base_url,params=params)

    if response.status_code == 200:
        data = response.json()
        studies = data.get("StudyFieldsResponse", {}).get("StudyFields", [])
        
        for study in studies:
            print(f"NCT ID: {study.get('NCTId')}")
            print(f"Phase: {study.get('Phase')}")
            print(f"Start Date: {study.get('StartDate')}")
            print(f"Study Type: {study.get('StudyType')}")
            print(f"Condition: {study.get('Condition')}")
            print(f"Brief Title: {study.get('BriefTitle')}")
            print(f"EligibilityCriteria: {study.get('EligibilityCriteria')}")
            print("\n" + "-"*40 + "\n")

    else:
        print(f"Error: Unable to fetch data. Status code: {response.status_code}")

if __name__ == "__main__":
    search_term = input("Enter a search term for clinical trials: ")
    max_results = int(input("Enter the maximum number of results to retrieve: "))
    
    search_clinical_trials(search_term, max_results)
```
